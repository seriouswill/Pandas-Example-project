```python
# First lets tackle that typo!

df.rename(columns={df.columns[4]:"Appearances"}, inplace=True)

df.head()  # to check the name change

y = df.Appearances
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
footy_model = DecisionTreeRegressor(random_state=1)

# Fit model
footy_model.fit(x, y)
```

```python
print(df.Appearances.head())
print("Model predictions")
print(footy_model.predict(x.head()))

new_player_dict = {"Will `The Hitman` O'Connell":[["180", "0.42", "100"]],
                    "Jordan `The Darling` Darlington":[["102", "0.72", "100"]]}

key_list = list(new_player_dict.keys())
val_list = list(new_player_dict.values())

print(f"The matches neccesary for {key_list[0]} to score a hundred goals is {footy_model.predict(val_list[0])}!")

print(f"The matches neccesary for {key_list[1]} to score a hundred goals is {footy_model.predict(val_list[1])}!")
```
